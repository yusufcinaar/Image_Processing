```python
import cv2 
import numpy
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_original = cv2.imread("./image.jpg",1)
img_rgb=cv2.cvtColor(img_original,cv2.COLOR_BGR2RGB)
img_gray=cv2.cvtColor(img_original,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_original)
```

```python
fig = plt.figure(figsize=(10, 7))  # fig tanımlandı
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(img_original)
plt.title("BGR IMAGE")
plt.axis("off")

fig.add_subplot(rows, columns, 2)
plt.imshow(img_rgb)
plt.title("RGB Image")
plt.axis("off")

fig.add_subplot(rows, columns, 3)
plt.imshow(img_gray, cmap="gray")
plt.title("Gray Image")
plt.axis("off")

plt.show()
```

```python
height, width = img_original.shape[:2]

print(f"Görüntü Yüksekliği: {height}, Görüntü Genişliği: {width}")
```

```python
if height < 600 and width < 600:
    
    img_yarı = cv2.resize(img_original, (width // 2, height // 2), interpolation=cv2.INTER_AREA)
    

    img_katı = cv2.resize(img_original, (width * 2, height * 2), interpolation=cv2.INTER_LINEAR)
```

```python
    plt.figure(figsize=(15, 5))

    # Orijinal görüntü
    plt.subplot(1, 3, 1)
    plt.imshow(img_original)
    plt.title(f"Orijinal\n{width} x {height}")
    plt.axis("off")

    # Yarısına küçültülmüş görüntü
    plt.subplot(1, 3, 2)
    plt.imshow(img_yarı)
    plt.title(f"Yarısı\n{width // 2} x {height // 2}")
    plt.axis("off")

    # 2 kat büyütülmüş görüntü
    plt.subplot(1, 3, 3)
    plt.imshow(img_katı)
    plt.title(f"2 Katı\n{width * 2} x {height * 2}")
    plt.axis("off")

    # Görüntüleri ekranda göster
    plt.show()
```

```python
flag =np.zeros((300, 600, 3), dtype=np.uint8)
```

```python
# Renkleri tanımla
flag[:100, :] = [0, 0, 0]         # Siyah şerit
flag[100:200, :] = [0, 0, 255]    # Kırmızı şerit
flag[200:, :] = [0, 255, 255]     # Sarı şerit
```

```python
# Bayrağı göster
plt.imshow(cv2.cvtColor(flag, cv2.COLOR_BGR2RGB))
plt.title("Almanya Bayrağı - 300x600")
plt.axis("off")  # Eksenleri kapat
plt.show()
```
