```python
import cv2 
import numpy
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_original = cv2.imread("./peppers.jpg",1)
```

```python
plt.imshow(img_original)
```

```python
img_rgb=cv2.cvtColor(img_original,cv2.COLOR_BGR2RGB)
img_gray=cv2.cvtColor(img_original,cv2.COLOR_BGR2GRAY)
```

```python
fig = plt.figure(figsize=(10, 7))  # fig tanmland
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(img_original)
plt.title("BGR IMAGE")
plt.axis("off")

fig.add_subplot(rows, columns, 2)
plt.imshow(img_rgb)
plt.title("RGB Image")
plt.axis("off")

fig.add_subplot(rows, columns, 3)
plt.imshow(img_gray, cmap="gray")
plt.title("Gray Image")
plt.axis("off")

plt.show()
```

```python
img_rgb_indexes = (img_rgb[:, :, 0] > 130) & (img_rgb[:, :, 1] < 80) & (img_rgb[:, :, 2] < 70)
```

```python
new_rgb = cv2.cvtColor(img_original,cv2.COLOR_BGR2RGB)
```

```python
fig = plt.figure(figsize=(10, 7))  # fig tanmland
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(new_rgb)
plt.title("RGB IMAGe")
plt.axis("off")

fig.add_subplot(rows, columns, 2)
plt.imshow(img_rgb_indexes,cmap="gray")
plt.title("INDEXES Image")
plt.axis("off")

fig.add_subplot(rows, columns, 3)
plt.imshow(img_rgb,)
plt.axis("off")

plt.show()
```

```python
img_bgr = img_rgb  # orijinal RGB grntnz olabilir
```

```python
new_rgb_gray = cv2.cvtColor(new_rgb, cv2.COLOR_BGR2GRAY)
```

```python
hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
```

```python
lower_green = np.array([35, 40, 40])  # Alt snr (HSV formatnda yeilin daha geni aral)
upper_green = np.array([85, 255, 255])  # st snr

# Yeil renkler iin maske olutur
green_mask = cv2.inRange(hsv, lower_green, upper_green)

# Maske ile grntdeki yeil alanlar tespit edip pembe renge evirelim
new_rgb = img_bgr.copy()
new_rgb[green_mask > 0] = [235, 135, 235]  # Pembe renge evir
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 4

# Orijinal RGB Grnts
fig.add_subplot(rows, columns, 1)
plt.imshow(cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB))  # BGR'den RGB'ye dntr
plt.title("Original RGB Image")
plt.axis("off")

# Yeil Maskesi Grnts
fig.add_subplot(rows, columns, 2)
plt.imshow(green_mask, cmap='gray')
plt.title("Green Mask")
plt.axis("off")

# Maskeden nceki RGB Grnts
fig.add_subplot(rows, columns, 3)
plt.imshow(cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB))
plt.title("Original RGB (Before Mask)")
plt.axis("off")

# Maskeli RGB Grnts (Renk deitirilmi hali)
fig.add_subplot(rows, columns, 4)
plt.imshow(new_rgb_gray, cmap='gray')  # Gri tonlamal grnt
plt.title("Masked Grayscale Image")
plt.axis("off")

plt.show()
```

```python
height,width = 100,100

img_red = np.zeros((height,width,3),dtype=np.uint8)
img_red[:,:,0]=255

img_green = np.zeros((height,width,3),dtype=np.uint8)
img_green[:,:,1]=255

img_blue= np.zeros((height,width,3),dtype=np.uint8)
img_blue[:,:,2]=255

img_yellow = np.zeros((height, width, 3), dtype=np.uint8)
img_yellow[:, :, 0] = 255  # Krmz kanal
img_yellow[:, :, 1] = 255  # Yeil kanal


```
