```python
import cv2 
import numpy
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_original = cv2.imread("./peppers.jpg",1)
```

```python
plt.imshow(img_original)
```

```python
img_rgb=cv2.cvtColor(img_original,cv2.COLOR_BGR2RGB)
img_gray=cv2.cvtColor(img_original,cv2.COLOR_BGR2GRAY)
```

```python
fig = plt.figure(figsize=(10, 7))  # fig tanımlandı
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(img_original)
plt.title("BGR IMAGE")
plt.axis("off")

fig.add_subplot(rows, columns, 2)
plt.imshow(img_rgb)
plt.title("RGB Image")
plt.axis("off")

fig.add_subplot(rows, columns, 3)
plt.imshow(img_gray, cmap="gray")
plt.title("Gray Image")
plt.axis("off")

plt.show()
```

```python
new_rgb = cv2.cvtColor(img_original,cv2.COLOR_BGR2RGB)
```

```python
fig = plt.figure(figsize=(10, 7))  # fig tanımlandı
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(new_rgb)
plt.title("RGB IMAGe")
plt.axis("off")

fig.add_subplot(rows, columns, 2)
plt.imshow(img_rgb_indexes,cmap="gray")
plt.title("INDEXES Image")
plt.axis("off")

fig.add_subplot(rows, columns, 3)
plt.imshow(img_rgb,)
plt.axis("off")

plt.show()
```

```python
img_bgr = img_rgb  # orijinal RGB görüntünüz olabilir
```

```python
new_rgb_gray = cv2.cvtColor(new_rgb, cv2.COLOR_BGR2GRAY)
```

```python
hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
```

```python
lower_green = np.array([35, 40, 40])  # Alt sınır (HSV formatında yeşilin daha geniş aralığı)
upper_green = np.array([85, 255, 255])  # Üst sınır

# Yeşil renkler için maske oluştur
green_mask = cv2.inRange(hsv, lower_green, upper_green)

# Maske ile görüntüdeki yeşil alanları tespit edip pembe renge çevirelim
new_rgb = img_bgr.copy()
new_rgb[green_mask > 0] = [235, 135, 235]  # Pembe renge çevir
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 4

# Orijinal RGB Görüntüsü
fig.add_subplot(rows, columns, 1)
plt.imshow(cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB))  # BGR'den RGB'ye dönüştür
plt.title("Original RGB Image")
plt.axis("off")

# Yeşil Maskesi Görüntüsü
fig.add_subplot(rows, columns, 2)
plt.imshow(green_mask, cmap='gray')
plt.title("Green Mask")
plt.axis("off")

# Maskeden önceki RGB Görüntüsü
fig.add_subplot(rows, columns, 3)
plt.imshow(cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB))
plt.title("Original RGB (Before Mask)")
plt.axis("off")

# Maskeli RGB Görüntüsü (Renk değiştirilmiş hali)
fig.add_subplot(rows, columns, 4)
plt.imshow(new_rgb_gray, cmap='gray')  # Gri tonlamalı görüntü
plt.title("Masked Grayscale Image")
plt.axis("off")

plt.show()
```
