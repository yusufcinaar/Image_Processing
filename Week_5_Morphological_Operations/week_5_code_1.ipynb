```python
import cv2 
import numpy
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_image=cv2.imread("./1.png",0)
my_kernel=np.ones((5,5),dtype=np.uint8)
my_kernel
```

```python
my_erosion_img=cv2.erode(my_image,my_kernel,iterations=2)
fig=plt.figure(figsize=(10,7))
rows=1
columns=2
fig.add_subplot(rows,columns,1)
plt.imshow(my_image,cmap="gray")
plt.title("Orginal Image")
plt.axis("off")
fig.add_subplot(rows,columns,2)
plt.imshow(my_erosion_img,cmap="gray")
plt.title("Erosion")
plt.axis("off")
```

```python
new_img=cv2.imread("./2.png",0)

white_noise = np.random.randint(0, 2, size=new_img.shape[:2])
white_noise=white_noise * 255 

fig=plt.figure(figsize=(10,7))
rows=1
columns=2
fig.add_subplot(rows,columns,1)
plt.imshow(new_img,cmap="gray")
plt.title("Orginal Image")
plt.axis("off")

fig.add_subplot(rows,columns,2)
plt.imshow(white_noise,cmap="gray")
plt.title("white noise")
plt.axis("off")
```

```python
# Orijinal resmi oku
new_img = cv2.imread("./2.png", 0)

# Beyaz gürültü oluştur
white_noise = np.random.randint(0, 2, size=new_img.shape[:2])
white_noise = white_noise * 255 

# Beyaz gürültü ile orijinal resmi birleştir
combined_image = cv2.addWeighted(new_img, 0.5, white_noise.astype(np.uint8), 0.5, 0)

# Görselleri gösterme
fig = plt.figure(figsize=(15, 10))
rows = 1
columns = 3

# Orijinal görüntü
fig.add_subplot(rows, columns, 1)
plt.imshow(new_img, cmap="gray")
plt.title("Original Image")
plt.axis("off")

# Beyaz gürültü görüntüsü
fig.add_subplot(rows, columns, 2)
plt.imshow(white_noise, cmap="gray")
plt.title("White Noise")
plt.axis("off")

# Birleştirilmiş görüntü
fig.add_subplot(rows, columns, 3)
plt.imshow(combined_image, cmap="gray")
plt.title("Combined Image")
plt.axis("off")

plt.show()
```
